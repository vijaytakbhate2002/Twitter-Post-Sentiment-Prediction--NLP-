## NLP 

```python
import nltk
import re
from urllib.parse import urlparse
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings('ignore')

nltk.download('wordnet')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('omw-1.4')  
```

```python
lemmatizer = WordNetLemmatizer()
stop_words = list(stopwords.words('english'))
print(stop_words)
```

```python
from nltk import PorterStemmer
ps = PorterStemmer()

def stemmerAndLemmitization(sms:str):
        tokens = list(sms.split(' '))
        tokens = [ps.stem(word) for word in tokens]
        tokens = [word for word in tokens if word not in stop_words]
        tokens = [lemmatizer.lemmatize(word) for word in tokens]
        return tokens
    
def textProcess(sms):
    try:
        # 1. Remove URLs
        sms = re.sub(r'http\S+|www\S+|https\S+', '', sms, flags=re.MULTILINE)
        # 2. Remove mentions (@username) and hashtags (#hashtag)
        sms = re.sub(r'@\w+|#\w+', '', sms)
        # 3. Remove HTML tags
        sms = re.sub(r'<.*?>', '', sms)
        # 4. Remove emojis and special characters except alphanumeric
        # sms = re.sub(r'[^A-Za-z0-9\s]', '', sms)
        sms = re.sub(r'[^A-Za-z\s]', '', sms)
        sms = sms.lower()
        tokens = stemmerAndLemmitization(sms)
        processed_sms = ' '.join(tokens)
        return processed_sms

    except Exception as e:
        print(f"Couldn't process the input: {sms}")
        return "_None"

textProcess("Hey how are you today?, are you listening to me @musk there are 20million people")
```

```python
from tqdm import tqdm
tqdm.pandas()

train_df['tweets_processed'] = train_df['tweets'].progress_apply(lambda x: textProcess(x))
test_df['tweets_processed'] = test_df['tweets'].progress_apply(lambda x: textProcess(x))

df_irrelevant['tweets_processed'] = df_irrelevant['tweets'].progress_apply(lambda x: textProcess(x))
df_negative['tweets_processed'] = df_negative['tweets'].progress_apply(lambda x: textProcess(x))
df_neutral['tweets_processed'] = df_neutral['tweets'].progress_apply(lambda x: textProcess(x))
df_positive['tweets_processed'] = df_positive['tweets'].progress_apply(lambda x: textProcess(x))
```

## Vectorizing words from tweets
* TF-IDF (term frequency - inverse document frequency)

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split,cross_val_score

X = pd.concat([train_df['tweets_processed'], test_df['tweets_processed']], axis='rows')

tf_vec = TfidfVectorizer(min_df=0.001)
print("vector after instantiation ", tf_vec)
tf_vec_class = tf_vec.fit_transform(X)
print("vector after transformation ", tf_vec_class)
tf_vec_arr = tf_vec_class.toarray()
print("vector after conversion into array  ", tf_vec_arr)
df_vec = pd.DataFrame(tf_vec_arr, columns=tf_vec.get_feature_names_out())

df_vec.head()
```

## Model training

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
from sklearn.linear_model import LogisticRegression
```

```python
y = pd.concat([train_df['@sentiments'], test_df['@sentiments']], axis='rows')

X_train,X_test,y_train,y_test = train_test_split(df_vec, y, stratify=y, test_size=0.5)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## Data Saving

```python
class SampleIndices:
    def __init__(self, y):
        self.y = y

    def stratifiedIndices(self, sample_rate:float) -> dict:
        label_indices = []
        for label in self.y.unique():
            label_indices.append(self.y[self.y == label][:int(sample_rate * len(self.y[self.y == label]))].index.tolist())
        label_indices = [val for sublist in label_indices for val in sublist] 
        return label_indices

    def getSamples(self, sample_rate:float=0.7, stratify:bool=True) -> pd.DataFrame:
        """ Sample_rate: float range from 0.1 to 1.0, default=70
            statify: bool, default True """
        if stratify:
            label_indices = self.stratifiedIndices(sample_rate)
            return label_indices
        return self.y.index.tolist()[:int(sample_rate * len(self.y))]
        

sampler = SampleIndices(y=y)
indices = sampler.getSamples(sample_rate=0.2, stratify=True)
print(len(indices))
```

```python
df_vec_samples = df_vec.iloc[indices]
y_samples = y.iloc[indices]
```
